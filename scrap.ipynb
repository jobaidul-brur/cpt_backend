```python
from bs4 import BeautifulSoup
import requests


data = requests.get("https://codeforces.com/problemset/problem/1886/F").content
```

```python

soup = BeautifulSoup(data, 'html.parser')


# Extract relevant information from the soup object
problem_title = soup.select_one('.problem-statement .title').text.strip()


problem_description = soup.select_one('.problem-statement').findChildren('div', recursive=False)[1]
input_constraints = soup.select_one('.problem-statement .input-specification')
output_constraints = soup.select_one('.problem-statement .output-specification')
sample_tests = soup.select_one('.problem-statement .sample-test')

# Return the scraped data as a dictionary or store it in your Django models
print(soup)
problem_data = {
    'title': problem_title,
    'description': problem_description,
}
```
